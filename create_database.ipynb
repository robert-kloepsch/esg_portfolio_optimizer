```python
df = yf.download(stocks, start="2019-12-30", end="2020-12-30")
df = df["Adj Close"].reset_index()
```

```python
new = pd.DataFrame(columns=["pos_id", "sim_valuationdate", "isin", "market_price", "cia_rating", "sector", "continent", "ccy", "company"])
new = pd.DataFrame(columns=["date", "market_price", "company", "pos_id", "isin", "currency", "sector", "cia_rating", "continent"])
date_list = df["Date"].to_list()
```

```python
new = pd.DataFrame(columns=["pos_id", "sim_valuationdate", "isin", "market_price", "cia_rating", "sector", "continent", "ccy", "company"])
date_list = df["Date"].to_list()

for index, column in enumerate(df.drop("Date", axis=1)):
    random_esg_score = random.randint(5, 12)
    random_continent = random.randint(0, 4)
    for index_2, price in enumerate(df[column]):
        df_temp = pd.DataFrame({"pos_id": [isin_list[index]],  
                                "sim_valuationdate": [date_list[index_2]],
                                "isin": [isin_list[index]],
                                "market_price": [price],
                                "cia_rating": [random_esg_score],
                                "sector": [sector[index]],
                                "continent": [continents[random_continent]],
                                "ccy": [currency_list[index]],
                                "company": [stocks[index]]})
        new = pd.concat([new, df_temp])
```
